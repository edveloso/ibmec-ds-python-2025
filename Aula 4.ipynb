```python
import re
def limpa_titulo(titulo:str):
    return re.split(r'\(|-',titulo)[0].strip()
dados ['Main Job'] = dados ['Job Title'].apply (limpa_titulo)

def converte_data (data_str:str):
    if 'd' in data_str:
        return int (data_str.replace('d','').replace('+','').strip())
    if 'h' in data_str:
        return int (data_str.replace('h','').replace('+','').strip())/24
    if 'w' in data_str:
        return int (data_str.replace('w','').replace('+','').strip())*7
    else:
        return None
    
dados ['Days Post'] = dados ['Date'].apply(converte_data)

def converte_salario (salario_str):
    salario_range = re.findall (r'\$([\dK])', salario_str)
    salario_range = [int(s.replace ('K',''))*1000 for s in salario_range]
    if len (salario_range)==2:
        return sum (salario_range)/2
    else:
        return None
    
dados ['Salary'] = dados ['Salary'].astype(str)
dados ['Salario Medio']=dados['Salary'].apply(converte_salario)

dados
    
```
